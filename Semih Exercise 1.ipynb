# Exercise 1.1: Radial light rays in Schwarzschild spacetime

This sample code demonstrates how to numerically integrate an ordinary differential equation in Python.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
```

```python
def integrate_sample_ode(r0):
    t_history = [ 0. ]
    r_history = [ r0 ]
    r = ode(r_dot).set_integrator('zvode', method='bdf').set_initial_value(r_history[0], t_history[0])
    dt = 0.1
    while r.successful() and r.t < 10.:
        r_history.append(r.integrate(r.t + dt)[0])
        t_history.append(r.t)
    return t_history, r_history
```

```python
def r_dot(t,r) : 
    return (1-2/r)/(1+2/r)
```

```python
plt.figure(figsize=(9, 6))
for r0 in np.linspace(1., 3., 10):
    t, r = integrate_sample_ode(r0=r0)
    plt.plot(t, r, marker='.', ls='dotted')
```

# Exercise 1.1: Radial light rays in Schwarzschild spacetime

This sample code demonstrates how to numerically integrate an ordinary differential equation in Python.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
```

```python
def y_prime(x, y):
    return (1-(2/y))/(1+(2/y))
```

$y'=(1-(2/y(x,y)))/(1+(2/y(x,y)))$

```python
def integrate_sample_ode(x0):
    t_history = [ 0. ]
    x_history = [ x0 ]
    x = ode(y_prime).set_integrator('zvode', method='bdf').set_initial_value(x_history[0], t_history[0])
    dt = 0.1
    while x.successful() and x.t < 3.:
        x_history.append(x.integrate(x.t + dt)[0])
        t_history.append(x.t)
    return t_history, x_history
```

```python
plt.figure(figsize=(9, 6))
for x0 in np.linspace(0.1, 1., 10):
    t, x = integrate_sample_ode(x0=x0)
    plt.plot(t, x, marker='.', ls='dotted')
    
plt.xlabel("t")
plt.ylabel("x")
```
